### Load data, create atlas object

```python
dset_dir = '/mnt/cube/Datasets/ATLAS_starling/'

starling_atlas = atlas(
    dset_dir = dset_dir,
    um_mult = 100,
    smoothing = ['Brain', 'Brainregions', 'Nuclei']

)
```

```python
plot_transection(starling_atlas, regions_to_plot='Brainregions')
```
